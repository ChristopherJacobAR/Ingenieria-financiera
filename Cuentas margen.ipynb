## 1. Repaso: Bolsa de Valores y Mercados Secundarios

La bolsa de valores es una organización financiera que facilita y garantiza las transacciones de activos financieros, como futuros y opciones. Los mercados secundarios permiten la compra y venta de activos financieros previamente emitidos.

### 1.1 Diferencia entre Bolsa de Valores y Mercados Secundarios
- **Bolsa de Valores**: Lugar donde se emiten y negocian activos financieros como acciones, bonos y derivados.
- **Mercado Secundario**: Espacio donde se intercambian activos previamente emitidos entre inversores.

## 2. Participantes en los Mercados de Futuros
Existen tres tipos principales de participantes en los mercados de futuros:
- **Coberturistas (Hedgers)**: Reducen el riesgo relacionado con los movimientos de precios futuros.
- **Especuladores (Speculators)**: Apuestan por la dirección futura de los precios para obtener beneficios.
- **Arbitrajistas (Arbitrageurs)**: Buscan ganancias libres de riesgo a través de diferencias de precios en diferentes mercados.



## 2.1 Hedging (Cobertura de Riesgos)
El objetivo de un coberturista es minimizar el riesgo de fluctuaciones en los precios de los activos subyacentes. Por ejemplo, un agricultor puede utilizar futuros sobre maíz para protegerse contra la caída de los precios antes de la cosecha.

## Ejemplo de Cobertura con Contratos de Futuros

Un ejemplo clásico de cómo los **coberturistas** utilizan los contratos de futuros es el caso de un agricultor que cultiva maíz y desea protegerse contra la caída de los precios antes de la cosecha. Para evitar el riesgo de que el precio del maíz baje en el futuro, el agricultor puede vender contratos de futuros sobre maíz. Si el precio del maíz cae al momento de la cosecha, las pérdidas en el mercado físico serán compensadas por las ganancias obtenidas en el mercado de futuros.

### Ejemplo numérico:

Supongamos que un agricultor estima que cosechará **10,000 bushels de maíz** en tres meses. Actualmente, el precio de los futuros de maíz es de **$4.00 por bushel**. Para cubrirse, el agricultor vende futuros por **10,000 bushels** a ese precio.

Si dentro de tres meses el precio del maíz baja a **3.50 por bushel**, aunque el agricultor venda su maiz a un precio mas bajo en el mercado fisico, obtendra una **ganancia de 0.50 por bushel** en el mercado de futuros, compensando así la pérdida.

A continuación, veamos un ejemplo práctico en Python para calcular el resultado:




```python
# Ejemplo de liquidación de contratos de futuros mediante la compra de un contrato opuesto

# Supongamos que un participante compra un contrato de futuros para el maíz a $4.00 por bushel
precio_compra = 4.00  # Precio de compra en USD por bushel
cantidad_bushels = 10000  # Cantidad de bushels en el contrato

# Ahora el participante vende un contrato opuesto cuando el precio sube a $4.50 por bushel
precio_venta = 4.50  # Precio de venta en USD por bushel

# Cálculo de la ganancia o pérdida
ganancia_por_bushel = precio_venta - precio_compra
ganancia_total = ganancia_por_bushel * cantidad_bushels

print(f"Ganancia total al cerrar la posición: ${ganancia_total:.2f}")
```

## 4. Funcionamiento de las Cuentas de Margen

En los contratos de futuros, es necesario mantener cuentas de margen para garantizar que ambas partes cumplirán con sus obligaciones. Hay tres tipos de márgenes:
- **Margen Inicial**: Depósito inicial que se debe realizar al abrir una posición.
- **Margen de Mantenimiento**: Cantidad mínima que debe mantenerse en la cuenta.
- **Margen de Variación**: Ajustes diarios en función de las variaciones en el precio del activo subyacente.

### Ejemplo de Cálculo de Margin Call

Supón que compras dos contratos a futuro sobre jugo de naranja (cada contrato es por 15,000 libras) con un precio de 1.60 por libra. El margen inicial es de 6,000 por contrato y el margen de mantenimiento es de 4,500. ¿Cuánto debe caer el precio spot para que se genere una margin call?


```python
# Ejemplo de Cálculo de Margin Call
precio_inicial = 1.60  # USD por libra
precio_caida = 1.40  # Nueva cotización en USD por libra
margen_inicial = 6000
margen_mantenimiento = 4500
libra_por_contrato = 15000
numero_contratos = 2

# Cálculo de la pérdida
perdida_por_libra = precio_inicial - precio_caida
perdida_total = perdida_por_libra * libra_por_contrato * numero_contratos
print(f"Pérdida total: ${perdida_total}")
```

## 5. Convergencia de los Precios Futuros y Spot

En los mercados de futuros, los precios de los contratos tienden a **converger** con el precio spot del activo subyacente a medida que se acerca la fecha de vencimiento. Esta convergencia está impulsada por las fuerzas de oferta y demanda y garantiza que no haya grandes diferencias entre los dos precios al vencimiento del contrato.

### Conceptos Importantes:
- **Contango**: Cuando el precio futuro es mayor que el precio spot.
- **Backwardation**: Cuando el precio futuro es menor que el precio spot.

Veamos cómo funcionan estos dos conceptos:



```python
import matplotlib.pyplot as plt

# Datos de ejemplo para ilustrar Contango y Backwardation
fechas = ['Mes 1', 'Mes 2', 'Mes 3', 'Mes 4', 'Mes 5',"Mes 6",'Mes 7', 'Mes 8', 'Mes 9']
precio_spot = [50, 50, 50, 50, 50, 50, 50, 50, 50]  # Precio spot constante
precio_futuro_contango = [55, 54, 53, 52, 51, 50, 49, 48, 47]  # Precio en contango
precio_futuro_backwardation = [44,45, 46, 47, 48, 49,50,51,52]  # Precio en backwardation

# Gráfico de Contango y Backwardation
plt.plot(fechas, precio_spot, label="Precio Spot", linestyle='--', color='black')
plt.plot(fechas, precio_futuro_contango, label="Precio Futuro (Contango)", color='blue')
plt.plot(fechas, precio_futuro_backwardation, label="Precio Futuro (Backwardation)", color='red')
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.title("Convergencia de Precios Futuros y Spot")
plt.legend()
plt.grid(True)
plt.show()
```

## Ejemplo real con precios del petroleo

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Descargar datos de futuros del petróleo crudo (CL=F) desde 2015 hasta 2024
futuro = yf.download('CL=F', start='2008-01-01', end='2024-01-01', interval='1mo')

# Simulamos un precio spot constante (puedes usar un valor promedio del precio del futuro)
precio_spot = [futuro['Adj Close'].mean()] * len(futuro)  # Precio spot constante (promedio de los futuros)

# Simulamos contango y backwardation
precio_futuro_contango = futuro['Adj Close'] + 5  # Contango (precio futuro mayor al spot)
precio_futuro_backwardation = futuro['Adj Close'] - 5  # Backwardation (precio futuro menor al spot)

# Fechas de los datos
fechas = futuro.index.strftime("%Y-%m")

# Gráfico de Contango y Backwardation
plt.figure(figsize=(10, 6))
plt.plot(fechas, precio_spot, label="Precio Spot", linestyle='--', color='black')
plt.plot(fechas, precio_futuro_contango, label="Precio Futuro (Contango)", color='blue')
plt.plot(fechas, precio_futuro_backwardation, label="Precio Futuro (Backwardation)", color='red')
plt.xlabel("Fecha")
plt.ylabel("Precio (USD)")
plt.title("Convergencia de Precios Futuros y Spot")
plt.xticks(rotation=45, ticks=range(0, len(fechas), 12))  # Mostrar solo una fecha por año
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

En el gráfico anterior, se puede observar cómo en un escenario de Contango, el precio futuro comienza por encima del precio spot y tiende a converger hacia este. En un escenario de Backwardation, el precio futuro comienza por debajo del precio spot y también converge hacia este.

## 6. Ejercicio y Liquidación de Contratos

Los contratos de futuros generalmente se cierran antes de la fecha de entrega a través de una **liquidación en efectivo** o mediante la adquisición de un **contrato opuesto**. Esto evita la entrega física del activo subyacente.

### Pregunta:
**¿Por qué los participantes suelen cerrar sus posiciones antes de la entrega física?**

1. Para evitar los costos asociados con el **almacenamiento** o **manejo** del activo.
2. Porque en muchos casos, los participantes buscan **ganancias especulativas** en lugar de la adquisición física del activo.

Este comportamiento es común en los mercados de futuros, donde la mayoría de los contratos no llegan a la entrega física, ya que se utilizan principalmente para la especulación o la cobertura de riesgo sin intención de adquirir o entregar el activo subyacente.

## Ejercicio 2.11: Contrato de Futuros sobre Jugo de Naranja

Un negociante compra dos contratos (long) a futuro para el mes de Julio sobre el jugo de naranja. Cada contrato es por la entrega de **15,000 libras**. El precio actual a futuro es de **160 centavos (1.60 USD) por libra**, el margen inicial es de **6,000 por contrato** y el margen de mantenimiento es de **4,500 por contrato**.

### Preguntas:

1. ¿Cuánto debe caer el precio spot por libra para que se haga una petición de cobertura complementaria (margin call)? Respuesta en USD a dos decimales.
2. ¿De cuánto sería el margin call si el precio spot por libra cae a **1.40 USD**? (Recuerda son dos contratos).
3. ¿Qué precio spot por libra lleva a que se pueda retirar **3,000** de la cuenta margen? Respuesta en USD con dos decimales. (Recuerda son dos contratos).

### Solución:

#### 1. Cálculo del precio spot para una margin call:
El **margen inicial** es de **6000** por contrato y el **margen de mantenimiento** es de **$4500** por contrato. La diferencia es de **1500**, lo que significa que el contrato puede bajar hasta un total de **1500** antes de generar una margin call. 

#### 2. Cálculo del margin call si el precio cae a $1.40 USD:
Si el precio cae a **1.40 USD**, debemos calcular cuánto se ha reducido el valor del contrato y determinar el margin call correspondiente.



```python
precio_inicial = 1.60  # Precio inicial en USD por libra
precio_cae = 1.40  # Precio cuando ocurre margin call
margen_inicial = 6000  # Margen inicial por contrato
margen_mantenimiento = 4500  # Margen de mantenimiento por contrato
libras_por_contrato = 15000  # Cantidad de libras por contrato
num_contratos = 2  # Número de contratos

# Cálculo de la pérdida por libra
perdida_por_libra = precio_inicial - precio_cae

# Cálculo de la pérdida total por los dos contratos
perdida_total = perdida_por_libra * libras_por_contrato * num_contratos

# Cálculo del margin call
margin_call = perdida_total - ((margen_inicial - margen_mantenimiento) * num_contratos)
print(f"El margin call si el precio cae a $1.40 USD es: ${margin_call:.2f}")
```

### 3. Cálculo del precio spot para retirar $3,000 de la cuenta margen

Para determinar qué **precio spot** permite retirar **3,000** de la cuenta de margen, necesitamos calcular el aumento necesario en el valor del contrato. Esto implica calcular cuánto debe aumentar el precio spot por libra para que la cantidad de $3,000 pueda ser retirada.


```python
# Datos del problema
precio_inicial = 1.60  # Precio inicial en USD por libra
cantidad_a_retirar = 3000  # Cantidad a retirar de la cuenta en USD
libras_por_contrato = 15000  # Cantidad de libras por contrato
num_contratos = 2  # Número de contratos

# Cálculo del aumento en el precio necesario para retirar $3,000
precio_aumento = cantidad_a_retirar / (libras_por_contrato * num_contratos)

# Precio spot necesario para retirar $3,000
precio_necesario = precio_inicial + precio_aumento
print(f"El precio spot necesario para retirar $3,000 es: ${precio_necesario:.2f} por libra")
```

# Pregunta 2:
**¿De cuánto sería el margin call si el precio spot por libra cae a 1.40 USD? (Recuerda son dos contratos).**



**Respuesta: \$6,000 USD.**

#### **Explicación Detallada:**

1. **Cálculo de la Pérdida Total al Precio de \$1.40 USD por Libra:**

   - **Caída en el Precio por Libra:**
     $$
     \text{Caída en el Precio} = \$1.60 - \$1.40 = \$0.20 \text{ por libra}.
     $$
   - **Cantidad Total de Libras (2 contratos):**
     $$
     \text{Cantidad Total de Libras} = 2 \text{ contratos} \times 15,000 \text{ libras/contrato} = 30,000 \text{ libras}.
     $$
   - **Pérdida Total:**
     $$
     \text{Pérdida Total} = \text{Caída en el Precio} \times \text{Cantidad Total de Libras} = \$0.20 \times 30,000 = \$6,000.
     $$

2. **Determinación del Saldo de la Cuenta Después de la Pérdida:**

   - **Margen Inicial Total:**
     $$
     \text{Margen Inicial Total} = 2 \text{ contratos} \times \$6,000 = \$12,000.
     $$
   - **Saldo Después de la Pérdida:**
     $$
     \text{Saldo Final} = \text{Margen Inicial Total} - \text{Pérdida Total} = \$12,000 - \$6,000 = \$6,000.
     $$

3. **Cálculo del Margin Call Necesario:**

   - **Margen de Mantenimiento Total:**
     $$
     \text{Margen de Mantenimiento Total} = 2 \text{ contratos} \times \$4,500 = \$9,000.
     $$
   - Dado que el saldo final de \$6,000 es menor que el margen de mantenimiento total de \$9,000, se requiere un margin call.
   - **Monto del Margin Call:**
     $$
     \text{Margin Call} = \text{Margen Inicial Total} - \text{Saldo Final} = \$12,000 - \$6,000 = \$6,000.
     $$


```python
# Parámetro adicional para la caída del precio
precio_caida = 1.40  # Precio del activo después de la caída
caida_en_el_precio_por_libra = precio_futuro-precio_caida
libras_totales = libras_por_contrato*numero_contratos
perdida_total = caida_en_el_precio_por_libra * libras_totales
margen_inicial_total=numero_contratos*perdida_total
saldo_final=margen_inicial_total-perdida_total
margen_de_mantenimiento_total=margen_mantenimiento*numero_contratos
# Verificar si la pérdida excede el margen
margin_call = margen_inicial_total - saldo_final
if margin_call > 0:
    print(f"Se requiere un margin call de: ${margin_call:.2f}")
else:
    print("No se requiere margin call.")
```
