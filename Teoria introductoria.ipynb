# Riesgos Financieros: Sistemáticos y Estructurales

El **riesgo financiero** es la posibilidad de que una inversión o una empresa experimente pérdidas debido a factores que afectan negativamente su desempeño financiero. Existen diferentes tipos de riesgos financieros, pero dos de los más relevantes son el **riesgo sistemático** y el **riesgo estructural**.

## 1. Riesgo Sistemático

El **riesgo sistemático** (también conocido como **riesgo de mercado**) es el riesgo que afecta a todo el mercado o una gran parte de este. No puede ser eliminado mediante la diversificación y es causado por factores externos al control de una empresa o sector, como:
- Cambios en las tasas de interés.
- Recesiones económicas.
- Guerras o inestabilidad política.

Este tipo de riesgo afecta a todas las empresas e industrias, y es **no diversificable**.

### Ejemplos de Factores de Riesgo Sistemático:
- Crisis financieras globales (Ej: la crisis del 2008).
- Cambios en la política monetaria de los bancos centrales.
- Acontecimientos geopolíticos que afectan los mercados internacionales.

---

## 2. Riesgo Estructural

El **riesgo estructural** (también llamado **riesgo no sistemático** o **riesgo específico**) está relacionado con una empresa o sector en particular. A diferencia del riesgo sistemático, este riesgo **puede ser diversificado** mediante la construcción de un portafolio con diferentes activos. Algunos ejemplos de riesgo estructural incluyen:
- Problemas internos de una empresa, como fallas de gestión.
- Cambios regulatorios específicos a una industria.
- Disminución en la demanda de un producto o servicio particular.

### Ejemplos de Factores de Riesgo Estructural:
- Escándalos corporativos o quiebras.
- Cambios en las regulaciones de un sector específico.
- Fallos en la cadena de suministro de una empresa.

---

## 3. Diferencias entre el Riesgo Sistemático y el Riesgo Estructural

| Característica               | Riesgo Sistemático                          | Riesgo Estructural                           |
|------------------------------|---------------------------------------------|---------------------------------------------|
| **Afecta a**                 | Todo el mercado o una gran parte del mercado| Sectores, empresas o industrias específicas |
| **Diversificación**           | No es posible eliminarlo con diversificación| Puede ser reducido mediante diversificación |
| **Ejemplos**                  | Recesiones económicas, cambios en tasas de interés | Quiebra de una empresa, cambios regulatorios |

---

## Visualización: Comparación de Riesgos




```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos para la visualización
x = np.linspace(1, 10, 10)  # Número de activos en el portafolio
riesgo_sistematico = np.full(10, 0.3)  # El riesgo sistemático permanece constante
riesgo_estructural = 0.5 / np.sqrt(x)  # El riesgo estructural disminuye con la diversificación

# Crear la gráfica
plt.plot(x, riesgo_sistematico, label='Riesgo Sistemático (No Diversificable)', color='r', linestyle='--')
plt.plot(x, riesgo_estructural, label='Riesgo Estructural (Diversificable)', color='b')

# Personalizar la gráfica
plt.title('Comparación de Riesgo Sistemático y Estructural')
plt.xlabel('Número de Activos en el Portafolio')
plt.ylabel('Nivel de Riesgo')
plt.legend()
plt.grid(True)
plt.show()
```

## Conceptos de Posicionamiento en Futuros y Forwards

### Posición Larga (Long)

El **comprador** de un contrato futuro o forward espera que el precio del activo suba, para beneficiarse de la diferencia entre el precio pactado y el precio de mercado.

### Posición Corta (Short)

El **vendedor** de un contrato futuro o forward espera que el precio del activo baje, para beneficiarse vendiendo a un precio más alto que el de mercado.

---
Las posiciones se pueden **cerrar** al realizar una transacción contraria a la inicial antes de la fecha de vencimiento del contrato.


```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de la posición larga y corta
precios = np.linspace(50, 150, 100)
precio_pactado = 100

# Beneficios para posición larga y corta
beneficio_larga = precios - precio_pactado  # Ganancia cuando el precio sube
beneficio_corta = precio_pactado - precios  # Ganancia cuando el precio baja

# Crear la visualización
plt.figure(figsize=(10, 6))

# Graficar la posición larga
plt.plot(precios, beneficio_larga, label='Posición Larga (Long)', color='green')

# Graficar la posición corta
plt.plot(precios, beneficio_corta, label='Posición Corta (Short)', color='red')

# Líneas de referencia
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(precio_pactado, color='black', linestyle='--')

# Añadir etiquetas y título
plt.title('Posiciones en Futuros y Forwards: Larga vs Corta', fontsize=14)
plt.xlabel('Precio del Activo', fontsize=12)
plt.ylabel('Beneficio', fontsize=12)
plt.grid(True)
plt.legend()

# Mostrar gráfico
plt.show()
```


## Opciones

Las **opciones** son contratos que otorgan al titular el derecho, pero no la obligación, de comprar o vender un activo en el futuro a un precio específico.

Las **opciones** son contratos que otorgan el derecho, pero no la obligación, de comprar (opciones de compra CALL) o vender (opciones de venta PUT) un activo subyacente a un precio específico antes o en una fecha determinada.

### Tipos de Opciones

- **Call (Opción de compra)**: Da el derecho a comprar un activo.
- **Put (Opción de venta)**: Da el derecho a vender un activo.

#### ***Las opciones pueden ser:***
- **Europeas**: Solo se pueden ejercer en la fecha de vencimiento.
- **Americanas**: Se pueden ejercer en cualquier momento antes de la fecha de vencimiento.

### Componentes de una Opción

- **S₀**: Precio actual del activo.
- **K**: Precio de ejercicio (strike price).
- **T**: Fecha de vencimiento.
- **C**: Precio de la opción de compra.
- **P**: Precio de la opción de venta.
- **r**: Tasa de interés libre de riesgo.



# FUTUROS 

Un **futuro** es un contrato que obliga a las partes a comprar o vender un activo en una fecha futura a un precio acordado. Los futuros se negocian en mercados organizados como la **Bolsa Mercantil de Chicago (CME)**.

### Ejemplo: Contrato a Futuro

Un **futuro** es un acuerdo para comprar o vender un activo en una fecha futura a un precio determinado. Los principales componentes de un contrato a futuro son:
- **S₀**: Precio actual.
- **F**: Precio futuro.
- **t**: Tiempo actual.
- **T**: Tiempo de ejercicio.

---
## 1.2 Mercados de Futuros

Los mercados de futuros fueron creados para ayudar a los productores y consumidores de materias primas a **gestionar el riesgo** de fluctuaciones en los precios. Un ejemplo típico es un agricultor que vende futuros para asegurar el precio de su cosecha.

| **Participante**    | **Descripción**                                                                         |
|---------------------|-----------------------------------------------------------------------------------------|
| **Coberturistas**    | Utilizan los futuros para protegerse contra el riesgo de cambios en los precios.         |
| **Especuladores**    | Intentan obtener beneficios prediciendo los movimientos de precios.                     |
| **Arbitrajistas**    | Aprovechan las discrepancias de precios entre diferentes mercados o productos financieros.|


## 1.3 Funcionamiento de un Contrato de Futuros

Un contrato de futuros incluye:
- **Activo subyacente**: El bien o activo financiero sobre el cual se basa el contrato (como petróleo, oro o índices bursátiles).
- **Tamaño del contrato**: La cantidad del activo que se va a comprar o vender.
- **Fecha de vencimiento**: La fecha en la que se deben cumplir las condiciones del contrato.
- **Precio de futuros**: El precio acordado para la transacción futura.
- **Entrega**: Las condiciones bajo las cuales se entregará el activo (si se llega a la entrega física).

La **cámara de compensación** garantiza que las transacciones se liquiden correctamente y minimiza el riesgo de incumplimiento.

Es importante que ambas partes del contrato comprendan estas especificaciones para evitar confusiones.

## 2.4 Precio de Futuros y Precio Spot

El **precio de futuros** refleja las expectativas sobre el valor futuro del activo subyacente. A medida que el contrato se acerca a su vencimiento, el precio de futuros converge hacia el **precio spot** (precio actual del activo en el mercado).

Si el precio de futuros es mayor que el spot, hay oportunidades de arbitraje vendiendo futuros y comprando el activo. Si el precio de futuros es menor que el spot, los inversionistas pueden comprar futuros y vender el activo.




```python
import matplotlib.pyplot as plt
import numpy as np

# Crear los datos
dias_hasta_vencimiento = np.linspace(0, 10, 100)  # Días hasta el vencimiento del contrato
precio_spot = 100  # Precio actual del activo en el mercado (spot)
precio_futuros = precio_spot + (20 * np.exp(-0.3 * dias_hasta_vencimiento))  # Precio de futuros converge hacia el spot

# Crear la gráfica
plt.figure(figsize=(10, 6))

# Graficar el precio de futuros
plt.plot(dias_hasta_vencimiento, precio_futuros, label='Precio de Futuros', color='blue', linestyle='--')

# Graficar el precio spot
plt.axhline(precio_spot, color='green', linestyle='-', label='Precio Spot')

# Añadir etiquetas y título
plt.title('Convergencia del Precio de Futuros hacia el Precio Spot', fontsize=14)
plt.xlabel('Días hasta el vencimiento', fontsize=12)
plt.ylabel('Precio del Activo', fontsize=12)
plt.legend()

# Mostrar gráfico
plt.grid(True)
plt.show()
```


## 2.6 Cotización y Liquidación

Los precios de los futuros se cotizan públicamente en las bolsas de valores y se ajustan continuamente durante las horas de negociación. Los contratos de futuros se liquidan de dos maneras:
1. **Liquidación antes del vencimiento**: Los participantes cierran sus posiciones realizando la transacción contraria.
2. **Liquidación a la entrega**: Si no se cierra la posición, se produce la entrega del activo subyacente o una liquidación en efectivo.

## 2.7 Cámaras de Compensación

Las cámaras de compensación actúan como intermediarias para asegurar que ambas partes del contrato cumplan con sus obligaciones. Estas instituciones reducen el riesgo de incumplimiento al garantizar el cumplimiento del contrato y administrar los márgenes.

## 2.8 Tipos de Órdenes

En el mercado de futuros se pueden usar diferentes tipos de órdenes, tales como:
| **Tipo de Orden**     | **Descripción**                                                                   |
|-----------------------|-----------------------------------------------------------------------------------|
| **Órdenes a mercado**  | Se ejecutan al mejor precio disponible.                                           |
| **Órdenes limitadas**  | Se ejecutan solo si el precio alcanza un nivel específico.                        |
| **Órdenes stop**       | Se activan cuando el precio alcanza un umbral predeterminado.                     |

## 1.4 Margen y Ajustes Diarios

Para asegurar el cumplimiento del contrato, las partes deben depositar un **margen inicial**. Este depósito se ajusta diariamente en función de las fluctuaciones del precio del contrato, lo que se conoce como **ajuste diario**. Si el saldo de la cuenta cae por debajo del **margen de mantenimiento**, se requiere un depósito adicional, llamado **margin call**.

## 2.5 Márgenes y Ajustes Diarios

El sistema de márgenes es fundamental para los contratos de futuros. Los participantes deben depositar un **margen inicial**, que es una fracción del valor total del contrato. Diariamente, el saldo de la cuenta se ajusta en función de las fluctuaciones del precio del contrato:
- Si el precio se mueve a favor del participante, se le abonan fondos.
- Si el precio se mueve en contra, debe aportar fondos adicionales.

El **margen de mantenimiento** es el monto mínimo que debe haber en la cuenta para mantener la posición abierta. Si la cuenta cae por debajo de este nivel, se realiza un **margin call**.


```python
import matplotlib.pyplot as plt
import numpy as np

# Simulación de días y precios de un contrato de futuros
dias = np.arange(0, 11)
precio_inicial = 100
precios = precio_inicial + np.random.randn(11) * 2  # Precios fluctuando

# Definir el margen inicial, margen de mantenimiento y balance
margen_inicial = 10
margen_mantenimiento = 7
balance_cuenta = np.zeros(len(dias))
balance_cuenta[0] = margen_inicial  # Empezamos con el margen inicial

# Simulación de ajuste diario del balance en función de las fluctuaciones de precios
ajuste_diario = precios - precio_inicial  # Cambios respecto al precio inicial

for i in range(1, len(dias)):
    balance_cuenta[i] = balance_cuenta[i-1] + ajuste_diario[i]
    if balance_cuenta[i] < margen_mantenimiento:
        balance_cuenta[i] = margen_inicial  # Margin Call (depósito adicional)

# Crear las gráficas
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

# Gráfico de los precios del contrato
ax[0].plot(dias, precios, label='Precio del contrato de futuros', color='blue', marker='o')
ax[0].axhline(precio_inicial, color='black', linestyle='--', label='Precio inicial ($100)')
ax[0].set_title('Fluctuación del Precio del Contrato de Futuros')
ax[0].set_xlabel('Días')
ax[0].set_ylabel('Precio del Contrato')
ax[0].legend()
ax[0].grid(True)

# Gráfico del balance de la cuenta ajustado diariamente
ax[1].plot(dias, balance_cuenta, label='Balance de la cuenta', color='green', marker='o')
ax[1].axhline(margen_inicial, color='black', linestyle='--', label='Margen Inicial ($10)')
ax[1].axhline(margen_mantenimiento, color='red', linestyle='--', label='Margen de Mantenimiento ($7)')
ax[1].set_title('Ajuste Diario del Balance de la Cuenta')
ax[1].set_xlabel('Días')
ax[1].set_ylabel('Balance de la Cuenta')
ax[1].legend()
ax[1].grid(True)

# Mostrar gráficos
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Definir los escenarios de precios de mercado
precios_mercado = [45, 57]

# Mariana y Leo: contratos a futuro
precio_futuro = 50  # Precio pactado a futuro

# Mariana y Leo con precios de mercado
def calcular_resultados(precio_futuro, precio_mercado):
    ganancia_mariana = precio_futuro - precio_mercado  # Ganancia o pérdida de Mariana
    ganancia_leo = precio_mercado - precio_futuro  # Ganancia o pérdida de Leo
    return ganancia_mariana, ganancia_leo

print("Resultados de Mariana y Leo:")
for precio in precios_mercado:
    mariana, leo = calcular_resultados(precio_futuro, precio)
    print(f"Precio de mercado: ${precio} -> Mariana: {mariana}, Leo: {leo}")

# Ejercicio: Contrato a Plazo vs Opción de Compra
precios_activo = list(range(800, 1300, 50))  # Precios del activo en un año
precio_ejercicio = 1000
prima_opcion = 100

# Contrato a Plazo: Utilidad es la diferencia entre el precio del activo y el precio pactado
def contrato_plazo(precio_activo, precio_ejercicio):
    return precio_activo - precio_ejercicio

# Opción de Compra: Utilidad es 0 si no se ejerce, o la diferencia menos la prima
def opcion_compra(precio_activo, precio_ejercicio, prima):
    if precio_activo > precio_ejercicio:
        return precio_activo - precio_ejercicio - prima
    else:
        return -prima

# Calcular utilidades para cada precio del activo
utilidad_plazo = [contrato_plazo(p, precio_ejercicio) for p in precios_activo]
utilidad_opcion = [opcion_compra(p, precio_ejercicio, prima_opcion) for p in precios_activo]

# Graficar los resultados
plt.plot(precios_activo, utilidad_plazo, label='Contrato a Plazo')
plt.plot(precios_activo, utilidad_opcion, label='Opción de Compra', linestyle='--')
plt.axhline(0, color='black',linewidth=0.5)
plt.title('Utilidad: Contrato a Plazo vs Opción de Compra')
plt.xlabel('Precio del Activo en un Año')
plt.ylabel('Utilidad')
plt.legend()
plt.grid(True)
plt.show()
```

# Ejercicio: Contrato a Plazo vs Opción de Compra

- Un negociante compra un activo en 1000 dentro de un año.
- Otro negociante compra una opción de compra (call) con el mismo precio de ejercicio de 1000 y paga una prima de 100.
- Diferencia en posiciones:
- El comprador del activo tiene una obligación de pagar 1000.
- El comprador de la opción solo ejercerá si el precio del activo es mayor a 1000.

## Descripción del Problema

En este ejercicio, comparamos las posiciones de dos negociantes:
- Un negociante compra un activo en **\$1000** mediante un contrato a plazo, el cual lo obliga a comprar el activo dentro de un año.
- Otro negociante compra una **opción de compra (call)** con el mismo precio de ejercicio de **\$1000**, y paga una prima de **\$100**.

La diferencia clave es que el comprador del contrato a plazo tiene la **obligación** de comprar el activo, mientras que el comprador de la opción de compra tiene el **derecho**, pero no la obligación, de comprar el activo si lo desea.

---

# Contrato a Plazo

En el contrato a plazo, la utilidad o pérdida para el comprador es la diferencia entre el precio de mercado del activo en un año **S_T**  y el precio acordado en el contrato  **K**.


$U = S_T - K$


Donde:
- \( S_T \): Precio del activo al final del año.
- \( K \): Precio de ejercicio o precio pactado del contrato (en este caso \$1000).

---

# Opción de Compra

Para la opción de compra, el comprador solo ejerce la opción si el precio del activo al final del año **S_T** es mayor que el precio de ejercicio **K**. La utilidad o pérdida  **U** para el comprador es la diferencia entre el precio de mercado y el precio de ejercicio, menos la **prima** que pagó por la opción.


\begin{cases} 
S_T - K - \text{Prima}, & \text{si } S_T > K \\
-\text{Prima}, & \text{si } S_T \leq K
\end{cases}


Donde:
- \( S_T \): Precio del activo al final del año.
- \( K \): Precio de ejercicio (\$1000).
- \( \text{Prima} \): Costo de la opción de compra (\$100).

---


## Diferencia en las Posiciones

- El comprador del **contrato a plazo** obtiene una utilidad o pérdida lineal basada en la diferencia entre el precio de mercado \( S_T \) y el precio pactado \( K \). Si \( S_T > K \), obtiene una ganancia, pero si \( S_T < K \), incurre en una pérdida.
- El comprador de la **opción de compra** tiene una pérdida limitada a la prima pagada si el precio \( S_T \leq K \), pero tiene un potencial de ganancia ilimitado si \( S_T > K \).

---

## Gráfico de Utilidad

A continuación, puedes generar un gráfico que muestre las utilidades del contrato a plazo y la opción de compra en función del precio del activo en un año (\( S_T \)).



```python
import matplotlib.pyplot as plt

# Datos del problema
precio_ejercicio = 1000  # Precio al que se comprará el activo
prima_opcion = 100  # Prima pagada por la opción de compra
precios_activo = list(range(800, 1300, 50))  # Precios posibles del activo en un año

# Función para calcular la utilidad del contrato a plazo
def contrato_plazo(precio_activo, precio_ejercicio):
    return precio_activo - precio_ejercicio

# Función para calcular la utilidad de la opción de compra
def opcion_compra(precio_activo, precio_ejercicio, prima):
    if precio_activo > precio_ejercicio:
        return precio_activo - precio_ejercicio - prima
    else:
        return -prima

# Calcular utilidades para cada precio del activo
utilidad_plazo = [contrato_plazo(p, precio_ejercicio) for p in precios_activo]
utilidad_opcion = [opcion_compra(p, precio_ejercicio, prima_opcion) for p in precios_activo]

# Graficar los resultados
plt.plot(precios_activo, utilidad_plazo, label='Contrato a Plazo', marker='o')
plt.plot(precios_activo, utilidad_opcion, label='Opción de Compra', linestyle='--', marker='x')
plt.axhline(0, color='black',linewidth=0.5)
plt.title('Utilidad: Contrato a Plazo vs Opción de Compra')
plt.xlabel('Precio del Activo en un Año')
plt.ylabel('Utilidad')
plt.legend()
plt.grid(True)
plt.show()
```

# Diferencias entre Forwards (Contratos a Plazo) y Futuros

Los **forwards** y los **futuros** son tipos de derivados financieros que permiten a los participantes del mercado comprar o vender un activo en una fecha futura a un precio determinado. Aunque comparten similitudes, existen diferencias clave entre ellos:

---

## 1. Mercado en el que se negocian

- **Forwards (Contratos a Plazo)**: 
  - Se negocian en mercados **no organizados** (Over-the-Counter, OTC).
  - Las condiciones del contrato (precio, vencimiento, etc.) son **personalizadas** entre las partes involucradas.
  
- **Futuros**: 
  - Se negocian en **mercados organizados** o bolsas de futuros, como la **Bolsa Mercantil de Chicago (CME)**.
  - Las condiciones del contrato están **estandarizadas** (activo subyacente, tamaño del contrato, fecha de vencimiento, etc.).

---

## 2. Liquidación

- **Forwards (Contratos a Plazo)**:
  - Generalmente, se **liquidan al vencimiento** mediante la entrega del activo subyacente, aunque también puede acordarse una liquidación en efectivo.
  
- **Futuros**:
  - La mayoría de los futuros se **liquidan diariamente** a través de la **cámara de compensación**, mediante ajustes diarios en función de las fluctuaciones del precio.
  - Pueden liquidarse antes de la fecha de vencimiento mediante la **compra o venta de una posición contraria**.

---

## 3. Riesgo de Contraparte

- **Forwards (Contratos a Plazo)**:
  - Existe un mayor riesgo de **incumplimiento** (default) por alguna de las partes, ya que no hay un intermediario que garantice el contrato.
  
- **Futuros**:
  - El riesgo de contraparte es prácticamente **eliminado** gracias a la **cámara de compensación**, que garantiza el cumplimiento de las obligaciones de ambas partes.

---

## 4. Ajuste Diario (Mark to Market)

- **Forwards (Contratos a Plazo)**:
  - No hay ajuste diario. Las ganancias o pérdidas se realizan en el momento de la liquidación del contrato, al vencimiento.
  
- **Futuros**:
  - Los futuros se ajustan diariamente mediante el **Mark to Market**: las ganancias o pérdidas diarias se abonan o deducen de la cuenta del participante.

---

## 5. Tamaño del Contrato

- **Forwards (Contratos a Plazo)**:
  - El tamaño del contrato es **flexible** y se puede adaptar a las necesidades de las partes.
  
- **Futuros**:
  - El tamaño del contrato está **estandarizado** y definido por la bolsa de futuros.

---

## 6. Uso Típico

- **Forwards (Contratos a Plazo)**:
  - Generalmente utilizados por empresas e instituciones para **cobertura** de activos específicos o transacciones futuras.
  
- **Futuros**:
  - Utilizados tanto para **cobertura** como para **especulación**. Los especuladores los usan para aprovechar los movimientos de precios en mercados líquidos.

---

### Resumen

| Característica                | Forwards                        | Futuros                          |
|-------------------------------|----------------------------------|----------------------------------|
| **Mercado**                    | OTC (No organizado)              | Mercado organizado (bolsa)       |
| **Liquidación**                | Al vencimiento                   | Ajustes diarios (Mark to Market) |
| **Riesgo de contraparte**      | Riesgo de incumplimiento         | Garantizado por cámara de compensación |
| **Ajuste Diario**              | No hay ajuste                    | Sí, mediante Mark to Market      |
| **Tamaño del contrato**        | Flexible (personalizable)        | Estandarizado                    |
| **Uso principal**              | Cobertura                        | Cobertura y especulación         |

---

En resumen, los **forwards** son más personalizados y se negocian de forma privada, mientras que los **futuros** son estandarizados, se negocian en bolsas y ofrecen mayor seguridad debido a la cámara de compensación y los ajustes diarios.
